```python
# Lecture du fichier CSV 'sample.csv'
df = pd.read_csv('/content/sample.csv')
# Affichage du contenu du DataFrame 'df', ce qui permet de visualiser les données du fichier CSV
#df
df.head()
```

**Lower casing**

```python
# Création d'une nouvelle colonne "text_lower" dans le DataFrame 'df'
# Contenant le contenu de la colonne "text" converti en lettres minuscules
df["text_lower"] = df["text"].str.lower()
df.head()
```

**Removal of Punctuations**

```python
PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    """function to remove the punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

df["text_no_punct"] = df["text_lower"].apply(lambda text: remove_punctuation(text))
df.head()
```

**Removal of Stopwords**

```python
# Importation de la bibliothèque nltk pour le traitement du langage naturel
import nltk
# Importation de la liste de mots vides (stopwords) depuis nltk
from nltk.corpus import stopwords
# Téléchargement de la liste de stopwords pour la langue anglaise
nltk.download('stopwords')
# Création d'un ensemble (set) de mots vides en anglais
STOPWORDS = set(stopwords.words('english'))
# Définition d'une fonction pour supprimer les mots vides du texte
def remove_stopwords(text):
    """Fonction pour supprimer les mots vides"""
    # Divise le texte en mots et conserve uniquement ceux qui ne sont pas dans STOPWORDS
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

# Application de la fonction remove_stopwords à la colonne 'text_no_punct' du DataFrame 'df'
# Résultat : création d'une nouvelle colonne 'text_no_stop_word' sans mots vides
df["text_no_stop_word"] = df["text_no_punct"].apply(lambda text: remove_stopwords(text))


df.head()
```

**Removal of Emojis**

```python
import re  # Importez le module 're' pour utiliser les expressions régulières
def remove_emojis(text):
    # Define a regular expression pattern for matching emojis
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F700-\U0001F77F"  # alchemical symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)

    # Remove emojis using the regex pattern
    return emoji_pattern.sub(r'', text)

df["text_no_emojis"] = df["text_no_stop_word"].apply(lambda text: remove_emojis(text))
#df.head()

# Affiche le contenu textuel du premier tweet dans la colonne "text_noçemojis" (index 0)
print(df['text_no_emojis'][0])
```

**Removal of URLs**

```python
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

df["text_no_urls"] = df["text_no_emojis"].apply(lambda text: remove_urls(text))
df.head()
```

**tokenization :**

```python
import spacy

# Chargez le modèle SpaCy pour le traitement du texte en anglais
nlp = spacy.load("en_core_web_sm")

# Fonction pour effectuer la tokenisation avec SpaCy
def tokenize_text(text):
    doc = nlp(text)
    # Récupérez les tokens (mots) du texte
    tokens = [token.text for token in doc]
    return tokens

# Appliquez la fonction de tokenisation à la colonne 'text_no_urls' du DataFrame
df["text_tokenization"] = df["text_no_urls"].apply(lambda text: tokenize_text(text))

df.head()
```

```python
print(df["text_tokenization"][0])
print(df["text_tokenization"][1])
```

**Lemmatization :**

```python
import spacy

# Chargement du modèle SpaCy pour le traitement du texte en anglais
nlp = spacy.load("en_core_web_sm")

# Définition d'une fonction pour lemmatiser les mots d'un texte
def lemmatize_words(text):
    # Analyse le texte avec le modèle SpaCy
    doc = nlp(text)
    # Récupère les lemmes (formes de base) de chaque token dans le texte
    lemmas = [token.lemma_ for token in doc]
    # Combine les lemmes en une seule chaîne de caractères avec des espaces
    return " ".join(lemmas)

# Application de la fonction de lemmatisation à la colonne 'text_no_urls' du DataFrame
df["text_lemmatized"] = df["text_no_urls"].apply(lambda text: lemmatize_words(text))

df.head()
```

```python
print(df["text_lemmatized"][0])
print(df["text_lemmatized"][1])
```

**Stemming :**

```python
from nltk.stem.porter import PorterStemmer
import nltk

# Téléchargez le modèle de tokenization Punkt
nltk.download('punkt')

stemmer = PorterStemmer()
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])

df["text_stemmed"] = df["text_no_urls"].apply(lambda text: stem_words(text))
df.head()
```
